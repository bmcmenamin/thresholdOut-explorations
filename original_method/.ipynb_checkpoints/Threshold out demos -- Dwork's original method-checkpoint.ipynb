## Replicating the original methods for holdout

This notebook uses python code based [Dworek *et al*'s 2015 report](http://science.sciencemag.org/content/349/6248/636) on the reusable holdout. Their code has been mirrored [here](https://github.com/andyljones/thresholdout)

```python
reps = 100
n, d = 5000, 500
krange = [0, 10, 20, 30, 45, 70, 100, 150, 200]
```

#### When there's no signal in the data
Here the top plot shows us that the standard holdout results in overfitting -- the performance on the holdout sample is greater than that of a fresh sample when using standard holdout. However, the bottom plot shows that using thresholdout results in performance on a holdout set that is comparable to a fresh set of data.

```python
original_demo.runandplotsummary(n, d, krange, reps, original_demo.createnosignaldata,  'nosignal')
```

#### When there a real signal in the data.

We see the same pattern here that we did in the no signal case -- the top plot (standard holdout) shows overfitting, and the bottom plot (thresholdout) shows supstantially reduced overfitting.

```python
# Experiment 2:
# Some variables are correlated

original_demo.runandplotsummary(n, d, krange, reps, original_demo.createhighsignaldata,  'highsignal')
```
